#### Drop  Dollar sign 

```python
df['Units Sold'] = df['Units Sold'].replace({'\$': '',',':'','\.': ''}, regex = True).astype(int)

df['Manufacturing Price'] = df['Manufacturing Price'].replace({'\$': '',',': '','\.': ''}, regex = True).astype(int)

df['Sale Price'] =df['Sale Price'].replace({'\$': '', ',':'', '\.':''}, regex = True).astype(int)

df['Gross Sales'] =df['Gross Sales'].replace({'\$': '', ',':'', '\.':''}, regex = True).astype(int)

df['Discounts'] = df['Discounts'].replace({'\$-': '0', '\$': '', ',': '', '\.': ''},regex = True).astype(int)

df['Sales'] =df['Sales'].replace({'\$': '', ',':'', '\.':''}, regex = True).astype(int)

df['COGS'] =df['COGS'].replace({'\$': '', ',':'', '\.':''}, regex = True).astype(int)

df['Profit'] = pd.to_numeric(df['Profit'].str.replace(r'[\$\,\.\-]', '', regex=True), errors='coerce')

df['Date'] =pd.to_datetime(df['Date'])
```

```python
df["Profit"].fillna(df['Profit'].mean(),inplace=True)
```

```python
sample = df['Profit']
```

```python
outliers =[]
def detect_outliers_iqr(data):
    data = sorted(data)
    q1 = np.percentile(data,25)
    q3 = np.percentile(data,75)
    IQR = q3 - q1
    lwr_bound = q1 - (1.5*IQR)
    upr_bound = q3 + (1.5*IQR)
    for i in data:
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    return outliers
sample_outliers = detect_outliers_iqr(sample)
print("Outliers from IQR method",sample_outliers)
```

```python
median = np.median(sample)
c=np.where(np.isin(sample,sample_outliers),median,sample)
print("Sample",sample)
print("New Array",c)
```

```python
df['Profit'] = c
```
